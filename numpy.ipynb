## Data Types

```python
#2^8 =128    -128 => 0 =>127

arr1 = np.array([1,3,2,4],dtype=np.int8)
#print(type(arr1))
print(arr1)
print(arr1.dtype)
print(f"{arr1.nbytes} bytes")
```

```python
#float datatype
arr2 = np.array([2.1,3.,3.8,6.6,7.0] ,dtype=np.float16)
print(arr2)
print(arr2.dtype)
print(f"{arr2.nbytes} bytes")
```

```python
#float datatype using int datatype
arr1 = np.array([100,232,234,543,323], dtype=np.float16)
print(arr1)
print(arr1.nbytes)
```

```python
#string datatype
arr1 = np.array([100,2432,234,543,323], dtype=np.str_)
print(arr1)
print(arr1.dtype)  #u4 show highest digit in an array U=>unique
print(f"{arr1.nbytes} bytes")

```

## Filtering

```python
arr=np.array([50,100,65,46,70])
print(arr[arr<=50])
p=arr[arr%2==0]
r=arr[(arr%2==0)|(arr<=50)]
print(p)
print(r)

arr =np.array([
    [1,2,3],
    [4,5,6]
])

even=arr[(arr%2==0) | ( arr==1)] # flatened single row changed
print(even)
even=np.where(arr%2==0 , arr ,0) #where condition is used to it can't be fattend
print(even)
```

## Random Numbers

## File Handling

```python
import os
print(os.getcwd())
import numpy as np

arr = np.array([1,2,3,4,5])
np.save("data.npy",arr)
read=np.load("data.npy")
print(read)

brr =np.array([6,7,8,9,0])
np.savez("data.npz" ,arr ,brr)
read=np.load("data.npz")
print(read)
print(read["arr_0"])
print(read["arr_1"])
```
