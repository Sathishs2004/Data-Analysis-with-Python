## Bar plot

```python
import numpy as np
sns.barplot(x='sex' , y='total_bill' , data=tip ,estimator=np.var)
```

```python
sns.countplot( x='sex' , data=tip) # count plot
```

## Box plot

```python
sns.boxplot(x='day' , y='total_bill' ,data=tip ,hue='sex')  
#center is median it show standard deviation
plt.legend(loc=0)
```

## violineplot

```python
sns.violinplot(x='day' , y='total_bill' ,data=tip ,hue='sex' ,split=True)  # as similar  as box plot  show distribution shape
# median outlayers distribution shape
```

## Swam plot

```python
sns.violinplot(x='day' , y='total_bill' ,data=tip ,hue='sex' ,split=True)
sns.swarmplot(x='day' , y='total_bill' ,data=tip ,hue='sex')
plt.legend(loc=0)
```

```python
sns.swarmplot(x='day' , y='total_bill' ,data=tip ,hue='sex')
plt.legend(loc=0)
```

## Strip plot

```python
sns.stripplot(x='day' , y='total_bill' ,data=tip ,jitter=True,hue='sex' ,dodge=True)
plt.legend(loc=0) # exact value can show it
```

## pallates

```python
sns.stripplot(x='day' , y='total_bill' ,data=tip ,jitter=True,hue='sex' ,dodge=True ,palette='Greens')
plt.legend(loc=1)
```
